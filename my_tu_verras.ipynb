```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
def load_dataset():
    # Replace 'path_to_csv_file' with the actual path to the CSV file
    url = 'https://storage.googleapis.com/qwasar-public/track-ds/boston.csv'
    data = pd.read_csv(url)
    return data

boston_data = load_dataset()
print(boston_data.head())
```

```python
def print_summarize_dataset(dataset):
    print("Dataset dimension:")
    print(f"{dataset.shape}")

    print("\nFirst few rows:")
    print(f"{dataset.head()}")

    print("\nStatistical summary:")
    print(f"{dataset.describe()}")

# Load the dataset
boston_data = load_dataset()

# Print the summary of the dataset
print_summarize_dataset(boston_data)
```

```python
def clean_dataset(boston_dataframe):
    # Remove rows with any missing values
    cleaned_data = boston_dataframe.dropna()
    return cleaned_data

# Load the dataset using pandas read_csv directly
url = 'https://storage.googleapis.com/qwasar-public/track-ds/boston.csv'
boston_data = pd.read_csv(url)

# Clean the dataset and store the result in cleaned_data
cleaned_data = clean_dataset(boston_data)

# Print the cleansed dataset
print(cleaned_data)

# Check if there are any missing values in the cleaned dataset
print(cleaned_data.isnull().sum())

# Now, the cleaned_data DataFrame should not contain any missing values
```

```python
def print_histograms(boston_dataframe):
    # Get the column names of the DataFrame
    columns = boston_dataframe.columns

    # Plot a histogram for each attribute
    for col in columns:
        plt.figure(figsize=(8, 6))
        plt.hist(boston_dataframe[col], bins=20, color='skyblue', edgecolor='black')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {col}')
        plt.grid(True)
        plt.show()

# Load the dataset
boston_data = load_dataset()

# Print histograms for all attributes
print_histograms(boston_data)
```

```python
def compute_correlations_matrix(boston_dataframe):
    # Compute the correlation matrix using pandas corr() method
    correlation_matrix = boston_dataframe.corr()

    return correlation_matrix

# Assuming you have already loaded the boston_data using the load_dataset() function
boston_data = load_dataset()

# Compute the correlation matrix
correlation_matrix = compute_correlations_matrix(boston_data)

# Print the shape of the correlation matrix
print(correlation_matrix.shape)
print(correlation_matrix['MDEV'])
```

```python
def print_scatter_matrix(boston_dataframe):
    # Create the scatter matrix using pandas.plotting.scatter_matrix
    scatter_matrix(boston_dataframe, alpha=0.8, figsize=(15, 15), diagonal='hist', grid=True)

    # Show the plot
    plt.show()

# Assuming you have already loaded the boston_data DataFrame using load_dataset()
boston_data = load_dataset()

# Call the function to print the scatter matrix
print_scatter_matrix(boston_data)
```

```python
import numpy as np # for computational anaylsis
import seaborn as sns 
from pandas.plotting import scatter_matrix
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# Prediction of model

def boston_fit_model(boston_dataframe):
    # Select two columns from the dataset: RM and MDEV
    model_dataset = boston_dataframe[["RM", "MDEV"]]
    regressor = LinearRegression()
    
    # Extract column 1 (RM) as input (X)
    x = model_dataset.iloc[:, :-1].values
    # Extract column 2 (MDEV) as output (y)
    y = model_dataset.iloc[:, 1].values
    
    # Train the model
    regressor.fit(x, y)
    return regressor

# Assuming you already have a DataFrame named 'boston_dataframe'
boston_dataframe = pd.read_csv('https://storage.googleapis.com/qwasar-public/track-ds/boston.csv')  
model = boston_fit_model(boston_dataframe)
print(model)
```
